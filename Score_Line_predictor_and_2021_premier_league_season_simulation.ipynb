## encoding

```python
# import pandas as pd

# Step 1: Calculate total goals scored by each team
team_goals = df_main_for_ordinal.groupby('home_team')['home_goals'].sum().reset_index()
away_team_goals = df_main_for_ordinal.groupby('away_team')['away_goals'].sum().reset_index()
team_goals['away_goals'] = away_team_goals['away_goals']

# Step 2: Sort the team_goals DataFrame by total goals
team_goals = team_goals.sort_values(by=['home_goals', 'away_goals'], ascending=False)

# Step 3: Create a dictionary to map team to ranking
team_ranking = {team: rank for rank, team in enumerate(team_goals['home_team'])}

# Step 4: Define custom ordinal encoding function
def custom_ordinal_encoding(x):
    return team_ranking.get(x, -1)

# Step 5: Apply custom ordinal encoding to home_team and away_team columns
df_main_for_ordinal['home_team_encoded'] = df_main_for_ordinal['home_team'].map(custom_ordinal_encoding)
df_main_for_ordinal['away_team_encoded'] = df_main_for_ordinal['away_team'].map(custom_ordinal_encoding)
```

```python
def distribute_ratings(formation, total_rating):
    # Split the formation string into a list of numbers
    positions = formation.split('-')

    # Create a list to store the number of players in each position
    num_players = [int(p) for p in positions]

    # Calculate the total number of players
    total_players = sum(num_players)

    # Calculate the percentage of the total rating for each position
    ratings = [total_rating * (num / total_players) for num in num_players]

    return ratings

position_ratings = distribute_ratings('3-5-2', 760)
```

## applying personal transformation where the tota player ratings are distributed into defence, mid, att by number of players in formation to both training and test(sim) data

### for training data


```python
# Iterate over each row in sim_data
for index, row in df1.iterrows():
    # Extract the formation and ratings for home team
    home_formation = row['formation']
    home_defense = row['defence rating']
    home_midfield = row['midfield rating']
    home_attack = row['attack rating']
    total_home_rating = home_defense + home_midfield + home_attack

    # Calculate the distributed ratings for home team
    home_ratings = distribute_ratings(home_formation, int(total_home_rating))
    home_defense_rating = home_ratings[0]
    home_midfield_rating = sum(home_ratings[1:-1])
    home_attack_rating = home_ratings[-1]

    # Extract the formation and ratings for away team
    away_formation = row['away formation']
    away_defense = row['away defence rating']
    away_midfield = row['away midfield rating']
    away_attack = row['away attack rating']
    total_away_rating = away_defense + away_midfield + away_attack

    # Calculate the distributed ratings for away team
    away_ratings = distribute_ratings(away_formation, total_away_rating)
    away_defense_rating = away_ratings[0]
    away_midfield_rating = sum(away_ratings[1:-1])
    away_attack_rating = away_ratings[-1]

    # Update the sim_data DataFrame with the redistributed ratings
    df1.at[index, 'defence rating'] = int(home_defense_rating)
    df1.at[index, 'midfield rating'] = int(home_midfield_rating)
    df1.at[index, 'attack rating'] = int(home_attack_rating)
    df1.at[index, 'away defence rating'] = int(away_defense_rating)
    df1.at[index, 'away midfield rating'] = int(away_midfield_rating)
    df1.at[index, 'away attack rating'] = int(away_attack_rating)
```

## model training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

weak_learner_final = DecisionTreeClassifier(max_depth=3)

adaBoost_clf_home_goals_final = AdaBoostClassifier(n_estimators=250, learning_rate = 0.08, random_state=0,base_estimator=weak_learner_final)
adaBoost_clf_away_goals_final = AdaBoostClassifier(n_estimators=250, learning_rate = 0.08, random_state=0,base_estimator=weak_learner_final)

adaBoost_clf_home_goals_final.fit(x_tree_based,y_home_goals)
adaBoost_clf_away_goals_final.fit(x_tree_based,y_away_goals)
```

# *SIMULATING*

```python

# Create a dictionary to store team information
teams = {}

# Create an empty list to store the fixtures results
fixtures_results = {}

# Iterate over each match
for index, row in df_sim.iterrows():
    home_team = row['home_team']
    away_team = row['away_team']

    # Check if the home team exists in the dictionary
    if home_team not in teams:
        teams[home_team] = {'points': 0, 'goals_scored': 0, 'goals_against': 0, 'games_played': 0}

    # Check if the away team exists in the dictionary
    if away_team not in teams:
        teams[away_team] = {'points': 0, 'goals_scored': 0, 'goals_against': 0, 'games_played': 0}

    # Prepare input matrix for home team prediction
    home_input = np.array([
        row['formation'], row['gk rating'], row['defence rating'], row['midfield rating'],
        row['attack rating'], row['net stamina'], row['net physicallity'],
        row['away formation'], row['away gk rating'], row['away defence rating'],
        row['away midfield rating'], row['away attack rating'], row['away net stamina'],
        row['away net physicallity'],
        row['home_team_encoded'], row['away_team_encoded']
    ]).reshape(1, -1)

    # Prepare input matrix for away team prediction
    away_input = np.array([
        row['formation'], row['gk rating'], row['defence rating'], row['midfield rating'],
        row['attack rating'], row['net stamina'], row['net physicallity'],
        row['away formation'], row['away gk rating'], row['away defence rating'],
        row['away midfield rating'], row['away attack rating'], row['away net stamina'],
        row['away net physicallity'],
        row['home_team_encoded'], row['away_team_encoded']
    ]).reshape(1, -1)

    # Predict goals for home team using goal_predictor_home model
    home_goals = adaBoost_clf_home_goals_final.predict(home_input)

    # Predict goals for away team using goal_predictor_away model
    away_goals = adaBoost_clf_away_goals_final.predict(away_input)

    # Update team statistics
    teams[home_team]['goals_scored'] += home_goals
    teams[home_team]['goals_against'] += away_goals
    teams[away_team]['goals_scored'] += away_goals
    teams[away_team]['goals_against'] += home_goals

    # Update games played
    teams[home_team]['games_played'] += 1
    teams[away_team]['games_played'] += 1

    # Determine the match result and update points
    if home_goals > away_goals:
        teams[home_team]['points'] += 3
    elif home_goals < away_goals:
        teams[away_team]['points'] += 3
    elif home_goals == away_goals:
        teams[home_team]['points'] += 1
        teams[away_team]['points'] += 1

    fixtures_results[index] = {
        'home_team': home_team,
        'away_team': away_team,
        'home_goals': home_goals[0],
        'away_goals': away_goals[0]
    }


# Create a DataFrame from the teams dictionary
team_table = pd.DataFrame.from_dict(teams, orient='index')
team_table.index.name = 'Team'

# Sort the table by points in descending order
team_table = team_table.sort_values(by='points', ascending=False)

print(team_table)
```
